```python
import pandas as pd
from time import sleep
```

```python
df = pd.read_excel(r'data\taxonomy Vocabs.xlsx')[['Variation', 'Is']]
df.head()
```

```python
taxonomy = {}
counter = 1
unique_parents = df['Variation'].unique().tolist()
for parent in unique_parents:
    print(f'{counter}: {parent}')
    children_df = df[df['Variation'] == f'{parent}'].dropna()
    children = children_df['Is'].unique().tolist()
    
    taxonomy[f'{parent}'] = children
    with open('data\taxonomy_final.txt', 'a+') as f:
        f.write(f'{parent}\n')
        for child in children:
            f.write(f'-{child}\n')
    
    counter += 1
    
taxonomy
```
