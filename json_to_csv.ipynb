```python
import os
import json
import csv
import re
import pandas as pd

df = pd.DataFrame()
for file in os.listdir('\data\ies_pub_files')[:100]:
    path = fr"{os.getcwd()}\data\ies_pub_files\\{file}"
#     print(path)
    with open(path, 'r') as f:
        df = df.append(pd.read_json(f, orient='index').T)

replace_dict = {
    re.compile(r'(\s)?\r'): ',',
    re.compile(r'  '): ' ',
    re.compile(r'\\xa0'): ' ',
    re.compile(r'(\s)?\\u.{4}(\s)?'): ' ',
    re.compile(r'\t+'): ' ',
    re.compile(r'•'): ''
}
df = df.replace(replace_dict, regex=True)
df.to_csv('data\site_json_1.csv', sep=",", quotechar='"', index=False, quoting=csv.QUOTE_ALL, encoding="utf-8-sig")
```
